```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=NKE&interval=5min&apikey=3UICUXH98811HV5J'
```

```python
r = requests.get(url)
data = r.json()
data = data['Time Series (5min)']
```

```python
dates = list(data.keys())[::-1]
prices = []
for date in dates:
    if '4. close' in data[date]:
        price = float(data[date]['4. close'])
        prices.append(price)
    

plt.plot(dates, prices)
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Price')

num_dates = len(dates)
if num_dates > 20:
    step = num_dates // 20 
    plt.xticks(range(0, num_dates, step), [date for i, date in enumerate(dates) if i % step == 0], rotation=45)

plt.show()
```
